# **Importing Essential Librarires**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import style
import warnings
import random
import sys
from scipy.stats import norm
warnings.filterwarnings('ignore')
style.use('ggplot')
from google.colab import files
random.seed(3)
```

# **Defining Functions for computing the D_x Model Value Theoretical and D_y Model Value Theoretical**

```python
def Dx_function(a,b,c,alpha,x,y,E_x):
  return_val = a*x + b*y + c + alpha*E_x
  return return_val
```

```python
def Dy_function(d,e,f,alpha,x,y,E_y):
  return_val = d*x + e*y + f + alpha*E_y
  return return_val
```

# **Main Function**

```python
# Load the data and covariance array
data = np.loadtxt('eddington_data.txt',skiprows=2)
x = data[:,0]
y = data[:,1]
E_x = data[:,2]
E_y = data[:,3]
Dx = data[:,4]
Dy = data[:,5]
sigma_Dx = 0.05
sigma_Dy = 0.05


# Likelihood function
def log_likelihood(params):
  alpha, a, b, c, d, e, f = params
  return_val = 0
  for i in range(len(data)):
    return_val = return_val + np.log(1/np.sqrt(2*np.pi*sigma_Dx**2)) + np.log(1/np.sqrt(2*np.pi*sigma_Dy**2))+ (-0.5 * (Dx[i] - Dx_function(a,b,c,alpha,x[i],y[i],E_x[i]))**2 * (1/(sigma_Dx*sigma_Dx))) + (-0.5 * (Dy[i] - Dy_function(d,e,f,alpha,x[i],y[i],E_y[i]))**2 * (1/(sigma_Dy*sigma_Dy)))
  return return_val

# Prior function
def log_prior(params):
    alpha, a, b, c, d, e, f = params
    if alpha>0:
        return 0.0
    else:
        return -np.inf

# Proposal function Gaussian
def propose(params, scale):
    return np.random.normal(params, scale)
```

# **MH - MCMC Code**

```python
# Define the Metropolis-Hastings algorithm
Rejected_Samples = []
def metropolis_hastings(log_likelihood, log_prior, propose, params, n_steps, scale):
    chain = [params]
    log_prob_chain = [log_prior(params) + log_likelihood(params)]
    n_accept = 0
    for i in range(n_steps):
        params_new = propose(params, scale)
        log_prob_new = log_prior(params_new) + log_likelihood(params_new)
        log_prob_old = log_prob_chain[-1]
        log_accept_ratio = log_prob_new - log_prob_old
        if np.log(np.random.rand()) < log_accept_ratio:
            chain.append(params_new)
            log_prob_chain.append(log_prob_new)
            params = params_new
            n_accept += 1
        else:
            Rejected_Samples.append((params_new))
            chain.append(params)
            log_prob_chain.append(log_prob_old)
    # # Discard the burn-in period of the chain
    burnin = int(n_steps * 0.3)
    return np.array(chain), np.array(log_prob_chain), n_accept/n_steps

# Set up initial parameters and run the MCMC algorithm
n_steps = 20000
scale = [0.001,0.01,0.1,1,10,100]
params0 = np.ones(7)
Acceptance = []
colors = ['red', 'green', 'blue', 'orange','purple', 'maroon']
for i in range(len(scale)):
  chain, log_prob_chain, acceptance_ratio = metropolis_hastings(log_likelihood, log_prior, propose, params0, n_steps, scale[i])
  print(str(i+1) + str(") "))
  print("")
  post_mean = np.mean(chain, axis=0)
  post_cov = np.cov(chain.T)

  # Plot the chains
  fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12,4))
  title_val = " Value of Alpha for Posterior Variance = " + str(scale[i])
  plt.title(title_val)
  plt.plot(chain[:,0], colors[i])
  plt.ylabel('Aplha')
  plt.xlabel('Number of iterations')
  plt.savefig(title_val + ".png")
  files.download(title_val + ".png")


  print('Mean of posterior distribution:', post_mean)
  print('Covariance of posterior distribution:', post_cov)
  print('Acceptance ratio:', acceptance_ratio)
  Acceptance.append(acceptance_ratio)
  print("")
```

```python
fig = plt.figure(figsize = (8,8) , dpi = 80)
plt.plot(scale, Acceptance , "g" , )
plt.xlabel("Scale")
plt.ylabel("Acceptance ratio")
plt.title("Acceptance Ratio verses Scale")
plt.xscale("log")
plt.savefig("Acceptance Ratio verses Scale.png")
files.download("Acceptance Ratio verses Scale.png")
```

**Thus, from the above calculated results we see that the following model converges to give the true values for the alpha at a posterior variance of 0.01 with mean alpha =  0.13903391.**
